# 計算指標並按照租金，售價匯總

1. 行政區售房單價元平方公尺 : district_sell
2. 行政區租房單價元平方公尺 : district_rent
3. 售租比 : sell2rentratio

```python
data_sell = df_alandtr_new[['district_id','unit_prc']].groupby(by = 'district_id').mean()
data_sell.rename(columns={'unit_prc':'district_sell'}, inplace = True)
data_sell.reset_index(drop=False,inplace = True)
data_rent = df_aaprrent[['district_id','unit_prc']].groupby(by = 'district_id').mean()
data_rent.rename(columns={'unit_prc':'district_rent'}, inplace = True)
data_rent.reset_index(drop=False,inplace = True)
data = pd.merge(data_rent,data_sell,left_on='district_id', right_on='district_id')
```

```python
data
```

```python
data['sell2rentratio'] = data['district_sell']/data['district_rent']
print('房屋租售比中位數為%i個月' % data['sell2rentratio'].median())
```

### One-Hot Encoding

One-Hot Encoding：這是一種常用的編碼方法，適用於無序的類別特徵。例如，如果你有一個特徵是城市名稱，包含'New York', 'San Francisco', 'Boston'等值，你可以使用One-Hot Encoding將這個特徵轉換為多個二元特徵。在Python中，你可以使用pandas的get_dummies函數來實現One-Hot **Encoding。**

Label Encoding：這種編碼方法將每個類別映射到一個整數。這種方法適用於有序的類別特徵，例如評分、等級等。在Python中，你可以使用sklearn的LabelEncoder來實現Label Encoding。

```python
import xgboost as xgb
import pandas as pd
from sklearn.metrics import mean_squared_error

# 假設你有一個名為 df 的 DataFrame，其中包含上述提到的特徵和目標變量
features = df[['build_area', 'room', 'district_rent', 'tot_floor', 'room_age','district',
        'rent_up_ll', 'coord_y', 'coord_x', 'cpi', 'revenue', 'day', 'month',
        'bstate_fg', 'district_sell']]
target = df['unit_prc']

# 分割訓練集和測試集
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# 設置 XGBoost 模型參數
params = {
    'n_estimators': 100,  # 樹的個數
    'max_depth': 3,  # 樹的最大深度
    'learning_rate': 0.1,  # 學習率
    'subsample': 0.8,  # 每棵樹樣本的子樣本比例
    'colsample_bytree': 0.8,  # 每棵樹特徵的子特徵比例
    'gamma': 0,  # 樹生長時需要的最小損失減少才會繼續分支
    'reg_alpha': 0,  # L1 正則化項的權重
    'reg_lambda': 1  # L2 正則化項的權重
}

# 建立 XGBoost 回歸器
model = xgb.XGBRegressor(**params)

# 訓練模型
model.fit(X_train, y_train)

# 使用模型進行預測
predictions = model.predict(X_test)

# # 顯示前10筆預測結果
for i in range(10):
    print(f"實際值: {y_test.iloc[i]}, 預測值: {predictions[i]}")

# 計算均方根誤差（RMSE）作為模型性能評估指標
rmse = mean_squared_error(y_test, predictions, squared=False)
# 該指標表示平均預測誤差的標準差。較小的RMSE值表示模型的預測性能較好。根據提供的RMSE值，可以說模型對目標變量的預測相對準確
print(f"均方根誤差（RMSE）: {rmse}")
```

## ARIMA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller

# Subset the relevant columns from the DataFrame
df_unit_prc = df_aaprrent_new[['mdate', 'unit_prc']]

# Set the 'date' column as the index
df_unit_prc.set_index('mdate', inplace=True)

# Check if the data is stationary and apply differencing if necessary
result = adfuller(df_unit_prc['unit_prc'])
if result[1] > 0.05:
    df_unit_prc_diff = df_unit_prc.diff().dropna()  # First order differencing
else:
    df_unit_prc_diff = df_unit_prc

# Plot the autocorrelation and partial autocorrelation functions
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8,8))
plot_acf(df_unit_prc_diff, ax=ax1)
plot_pacf(df_unit_prc_diff, ax=ax2)
plt.show()
```

```python
# Fit an ARIMA(1,1,1) model to the differenced data and obtain the model summary
model = ARIMA(df_unit_prc_diff, order=(1,1,1))
model_fit = model.fit()
print(model_fit.summary())
```

```python



# Make a forecast for the next 12 months
forecast = model_fit.forecast(steps=12)

# Convert the forecasted values to a DataFrame and add a 'date' column with the next 12 months of dates
forecast_df = pd.DataFrame(forecast[0], index=pd.date_range(start=df_unit_prc.index[-1], periods=12, freq='MS'), columns=['unit_prc'])

# Plot the original 'unit_prc' data and the forecasted values together
plt.plot(df_unit_prc, label='Original')
plt.plot(forecast_df, label='Forecast')
plt.legend()
plt.show()
```
